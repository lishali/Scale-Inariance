```python
#we will compare the asymtotics of the two functinos Phi and horizontal lattice, to see which gives better bounds as 
#s --> 1

def phi(s):
    return (np.pi)/2 - np.arcsin(1/s)

def Phi(s):
    return (2*np.ceil(np.pi/phi(s))*np.sqrt((1+np.ceil(1/(s+1)))*np.tan(phi(s))))/np.sin(phi(s))
#this is the same phi from theorem 5.2, i.e the one that uses superimposed rotated lattice estimates
    
def p(s):
    return np.ceil((2-s)/(s-1))
#the above is the number of horizonal lines per diamond

def optim(s):
    return np.sqrt(np.tan(phi(s))*(p(s)+2))
    
def horizontal_lattice(s):
    return 2*optim(s)/np.sin(phi(s))+p(s)*(1/optim(s))+optim(s)


print horizontal_lattice(1.5)
print (1/optim(1.4))
print p(2)
print optim(1.3)
```
