```python
# half of the minor angle of the diamond lattice given s
def phi(s):
    return (np.pi)/2 - np.arcsin(1/s)

# this is the Phi^{star} bound of 5.2, the one obtained via pi/phi_s superimposed rotations of the diamond lattice.  
def Phi(s):
    return (2*np.ceil(np.pi/phi(s))*np.sqrt((1+np.ceil(1/(s+1)))*np.tan(phi(s))))/np.sin(phi(s))

def p(s):
    return np.ceil((2-s)/(s-1))
#the above is the number of horizonal lines per diamond

#this is the optimized (minimized) y value given phi_s. Recall that y is the vertical height of each diamond (diamonds become flatter as s approaches 1)
def optim(s):
    return np.sqrt(np.tan(phi(s))*(p(s)+2))
    
#I meant to name this vertical lattice:
#we have the network value of the original diamond lattice (first term), second term is how many vertical root n lines we are adding
# finally the last term is adding lines of at most length y, for each pair of cities that are in the same diamond
#but have angle greater than phi_s (this is the angle in Lando's thesis

def horizontal_lattice(s):
    return 2*optim(s)/np.sin(phi(s))+p(s)*(1/optim(s))+optim(s)
```
