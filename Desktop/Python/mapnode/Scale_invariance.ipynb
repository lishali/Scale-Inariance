```python
from math import sin, cos, sqrt, atan2, radians

def E_distance(string1, string2):
    R = 6373.0
    lat1 = radians(float(string1[0]))
    lon1 = radians(float(string1[-1]))
    lat2 = radians(float(string2[0]))
    lon2 = radians(float(string2[-1]))
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

def reformat_string(string):
    try:
        a = string[1:-1]
            #b = str(sp['Location'][2*1+1])[1:-1]
        return a
    except Exception:
        print 'NaN'
        pass
    
def reformat_tuple(string):
    try:
        a = string[1:-1]
            #b = str(sp['Location'][2*1+1])[1:-1]
        a = [float(a.split(',')[0]), float(a.split(',')[-1])]
            #b = [float(b.split(',')[0]), float(b.split(',')[-1])]
        return a
    except Exception:
        print 'NaN'
        pass
    
    
print df['Location 1'][1]
reformat_tuple(df['Location 1'][1])

#except for the NaN entries, we are doing alright

```

```python
#And we are in business!  First the euclidean distance functions.  Since our data is a bit different this time, we need to redefine the e_distance functions

def e_distance(lat1, long1, lat2, long2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(long1)
    lat2 = radians(lat2)
    lon2 = radians(long2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

walmart['Euclidean Distance'] = walmart.apply(lambda x: e_distance(x['latitude'], x['longitude'], x['lat 2'], x['long 2']), axis=1)
```
